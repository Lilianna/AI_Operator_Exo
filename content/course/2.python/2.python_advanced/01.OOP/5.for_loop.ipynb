```python
""" Retrive and print words from an URL

Usage: 
    python3 words.py<URL>

"""

import sys
from urllib.request import urlopen

def fetch_words(url): 
    """"  Fetch a list of words from a URL
    
    Args: 
	    Url: the URL of a UTF-8 text document

    Returns: 
	    A list of strings containing the words from the document
    """

    story = urlopen(url)
    story_words = []
    for line in story:
        line_words = line.decode('utf8').split()
        for word in line_words:
            story_words.append(word)
    story.close()
    return story_words

def print_words(words):
    print("The story is read by line: ")
    for word in words:
        print(word)

def main():
    url = 'http://sixty-north.com/c/t.txt'
    ''' = sys.argv[1]'''
    
    words = fetch_words(url)
    print_words(words) 

if __name__ == '__main__':
    """Allowing to detect if a module is run as a script / imported into another module
    if dunder name is equal the string dunder name, we excute our function
    if not equal, the module knows it s imported into another module, 
    only define the function fetch_words, not excuting it 
    """
    '''main(sys.argv[1])'''
    main()

help(fetch_words(url))



```
